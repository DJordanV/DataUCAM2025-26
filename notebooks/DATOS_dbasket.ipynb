```python
import pandas as pd
import numpy as np

from pathlib import Path
```

```python
pd.options.display.max_rows = 100
pd.options.display.min_rows = 20
pd.options.display.max_columns = None
```

```python
ROOT_PATH = Path(Path.cwd()).resolve().parent
DATA_PATH = ROOT_PATH/'data'/'raw'/'dbasket'/'csv'
PROCESSED_PATH = ROOT_PATH/'data'/'processed'
```

# Creación de DataFrame principal

```python
dfs = []

for file in Path(DATA_PATH).glob('*.csv'):
    jornada = int(file.stem.split('_')[-1])
    df_temp = pd.read_csv(file)
    df_temp['Jornada'] = jornada
    dfs.append(df_temp)

df = pd.concat(dfs, ignore_index=True)
df = df.sort_values(by=['#', 'Jornada']).reset_index(drop=True)
```

# Modificaciones a DF

## Eliminar datos totales y de equipo

```python
df.dropna(subset=['#'], inplace=True)
```

## Cambio de nombre de columnas

```python
df = df.rename(columns= {'#ERROR!': '+/-'})
```

```python
cambio_dtype = {
    '#': 'int64',
    'TCA': 'int64',
    'TCI': 'int64',
    'T3A': 'int64',
    'T3I': 'int64',
    'TLA': 'int64',
    'TLI': 'int64',
    'ASI': 'int64',
    'ROB': 'int64',
    'TAP': 'int64',
    'TAPR': 'int64',
    'FPR': 'int64',
    'PTS': 'int64'
}
df = df.astype(cambio_dtype)
```
