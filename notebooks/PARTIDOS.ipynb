```python
df[['PTS_Loc', 'PTS_Vis']] = (
    df['Resultado']
    .str.split('-', expand=True)
    .apply(pd.to_numeric, errors='coerce')
    .fillna(0)
    .astype(int)
)

df['PTS_UCAM'] = np.where(df['Local'] == 'Local', df['PTS_Loc'], df['PTS_Vis'])
df['PTS_Adversario'] = np.where(df['Local'] == 'Local', df['PTS_Vis'], df['PTS_Loc'])

df['V/D'] = np.where(df['PTS_UCAM'] > df['PTS_Adversario'], 'VICTORIA', 'DERROTA')

df['Avg'] = df['PTS_UCAM'] - df['PTS_Adversario']
```

```python
df.dropna(inplace=True)
```
